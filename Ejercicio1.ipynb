## Problem 1 (Solving IVPs Using the Variation of Constants Formula)

Use the variation of constants formula from Theorem 1.6 to solve the following initial value problems of the form: $$ x' = m x + b , \quad x(t_0) = x_0 $$ where $m, b$ are constants.

---

### Step 1. Recognize the standard form and extract data

We start from the equation  
$x' = m\,x + b,$  
with $m,b\in\mathbb{R}$ constants, and initial condition  
$x(t_0)=x_0.$

1. **Linearity**. An equation of the form  
   $x' + a(t)\,x = b(t)$  
   is **first-order linear**.  
2. **Rewriting**. Here we have  
   $x' - \bigl(-m\,x\bigr) = b \quad\Longrightarrow\quad x' + \bigl(-m\bigr)x = b,$  
   but clearer is to write  
   $x' = m\,x + b \quad\Longrightarrow\quad x' - m\,x = b,$  
   or equivalently  
   $x' + (-m)\,x = b.$  
3. **Identification**:  
   $p(t)=m,\quad q(t)=b,\quad t_0 \text{ given},\quad x_0=x(t_0).$  
   Note that $p,q$ are continuous (they are constants).

---

### Step 2. Compute $\int_{t_0}^s p(\tau)\,d\tau$

For any continuous function $p(t)$, the **accumulated integral** from $t_0$ to $s$ is  
$\int_{t_0}^s p(\tau)\,d\tau.$

- In our case $p(\tau)=m$ is constant.  
- By the basic rule of integrals,  
  $\int_{t_0}^s m\,d\tau = m\,\int_{t_0}^s 1\,d\tau = m\,(s - t_0).$

This result measures the “area” under the curve $p(\tau)$ between $t_0$ and $s$, and is key for the integrating factor.

---

### Step 3. Formulate the integrand for the second part

The variation of constants formula presents the integrand  
$e^{-\int_{t_0}^s p(\tau)\,d\tau}\;q(s).$

- We substitute $\int_{t_0}^s p(\tau)\,d\tau = m(s - t_0).$  
- And $q(s)=b$, also constant.  
- Thus, $e^{-m(s - t_0)}\;b = b\,e^{-m(s - t_0)}.$

This factor $e^{-m(s-t_0)}$ “undoes” the growth that the homogeneous term $m\,x$ would introduce.

---

### Step 4. Evaluate the integral

$\int_{t_0}^t \bigl(e^{-\int_{t_0}^s p(\tau)\,d\tau}\,q(s)\bigr)\,ds
= \int_{t_0}^t b\,e^{-m(s - t_0)}\,ds.$

**Procedure**:

1. Factor out the constant $b$:  
   $b\int_{t_0}^t e^{-m(s - t_0)}\,ds.$  
2. Make the change of variable $u = s - t_0$, with $du = ds$.  
   - When $s=t_0$, $u=0$.  
   - When $s=t$,  $u=t - t_0$.  
3. The integral becomes  
   $b\int_{0}^{t - t_0} e^{-m\,u}\,du.$  
4. Compute  
   $\int e^{-m\,u}\,du = \frac{e^{-m\,u}}{-m},$  
   then  
   $b\left[\frac{e^{-m\,u}}{-m}\right]_0^{\,t - t_0}
   = \frac{b}{-m}\bigl(e^{-m(t - t_0)} - 1\bigr)
   = \frac{b}{m}\bigl(1 - e^{-m(t - t_0)}\bigr).$

This result is the contribution of the nonhomogeneous part.

---

### Step 5. Compute the factor $e^{\int_{t_0}^t p(\tau)\,d\tau}$

We need the full **integrating factor**:  
$\int_{t_0}^t p(\tau)\,d\tau = \int_{t_0}^t m\,d\tau = m\,(t - t_0),$  
and therefore  
$e^{\int_{t_0}^t p(\tau)\,d\tau} = e^{m(t - t_0)}.$

This factor multiplies both the initial condition and the integral from the previous step.

---

### Step 6. Assemble the solution according to formula (1.6)

The **variation of constants formula** gives  
$x(t) = e^{\int_{t_0}^t p(\tau)\,d\tau}\,x_0
\;+\;
e^{\int_{t_0}^t p(\tau)\,d\tau}
\;\int_{t_0}^t \bigl(e^{-\int_{t_0}^s p(\tau)\,d\tau}\,q(s)\bigr)\,ds.$

Substitute:

- $e^{\int_{t_0}^t p(\tau)\,d\tau}=e^{m(t - t_0)}.$  
- $\int_{t_0}^t \bigl(e^{-\int_{t_0}^s p(\tau)\,d\tau}q(s)\bigr)\,ds
   = \frac{b}{m}\bigl(1 - e^{-m(t - t_0)}\bigr).$

Thus,  
$x(t) = e^{m(t - t_0)}\,x_0
\;+\;
e^{m(t - t_0)}\;\frac{b}{m}\bigl(1 - e^{-m(t - t_0)}\bigr).$

---

### Step 7. Simplify the final expression

Expand the second term:  
$x(t)$  
$= x_0\,e^{m(t - t_0)} + \frac{b}{m}\bigl(e^{m(t - t_0)} - e^{m(t - t_0)}\,e^{-m(t - t_0)}\bigr)$  
$= x_0\,e^{m(t - t_0)} + \frac{b}{m}\bigl(e^{m(t - t_0)} - 1\bigr).$

To see its clear structure:

- **Homogeneous part**: $x_0\,e^{m(t - t_0)}$.  
- **Particular part**: $\frac{b}{m}\bigl(e^{m(t - t_0)} - 1\bigr)$.

Therefore, the most direct form is  
$x(t)
= x_0\,e^{m(t - t_0)}
\;+\;\frac{b}{m}\,\bigl(e^{m(t - t_0)} - 1\bigr).$

---

### Step 8. Rigorous verification

1. **Initial condition**. For $t=t_0$, $e^{m(t_0-t_0)}=1$, so  
   $x(t_0)
   = x_0\cdot1 + \frac{b}{m}(1-1)
   = x_0.$  

2. **Satisfaction of the differential equation**. Differentiate:  
   $x'(t)$  
   $= \frac{d}{dt}[x_0\,e^{m(t-t_0)}] + \frac{d}{dt}[\tfrac{b}{m}(e^{m(t-t_0)}-1)]$  
   $= x_0\,m\,e^{m(t-t_0)} + \frac{b}{m}\,m\,e^{m(t-t_0)}$  
   $= m(x_0 + \tfrac{b}{m})e^{m(t-t_0)}$  
   $= m\,x(t) + b.$

This demonstrates the **existence**, **uniqueness**, and **correctness** of the solution constructed by variation of constants.
