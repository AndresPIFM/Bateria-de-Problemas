## Problem 1 (Solving IVPs Using the Variation of Constants Formula)

$$
x' = -2\,x + \bigl(140 + 20e^{-2t}\bigr),\qquad x(0)=40,
$$

utilizando paso a paso la **fórmula de variación de constantes** (Teorema 1.6, ec. (1.5) del PDF) .

---

### Paso 1. Reconocer la forma estándar y garantizar existencia-única

1. **Forma lineal de primer orden.**  
   Toda ecuación de la forma  
   $$
     x' + p(t)\,x = q(t)
   $$  
   es lineal de primer orden. Aquí tenemos  
   $$
     x' = -2\,x + \bigl(140+20e^{-2t}\bigr)
     \quad\Longrightarrow\quad
     x' + 2\,x = 140 + 20e^{-2t}.
   $$
   De modo que identificamos  
   $$
     p(t)=2,\quad q(t)=140+20e^{-2t}.
   $$

2. **Condiciones iniciales.**  
   $$
     t_0=0,\quad x_0=40.
   $$

3. **Teorema de existencia y unicidad.**  
   Puesto que $p(t)$ y $q(t)$ son continuas en $\mathbb{R}$, el teorema de Cauchy–Lipschitz garantiza que existe una única solución $x(t)$ definida para todo $t \in \mathbb{R}$.

---

### Paso 2. Derivar y calcular el factor integrante

Para resolver  
$$
x' + p(t)\,x = q(t),
$$  
definimos el **factor integrante**  
$$
\mu(t)\;=\;\exp\Bigl(\int_{t_0}^{t}p(\tau)\,d\tau\Bigr).
$$  
Entonces, usando la regla del producto,

$$
\begin{aligned}
\frac{d}{dt}\bigl[\mu(t)\,x(t)\bigr]
&=\mu'(t)\,x(t)+\mu(t)\,x'(t)\\
&=\bigl(p(t)\mu(t)\bigr)x(t)+\mu(t)x'(t)\\
&=\mu(t)\bigl(x'+p\,x\bigr)\\
&=\mu(t)\,q(t).
\end{aligned}
$$

Integramos de $t_0$ a $t$:

$$
\mu(t)\,x(t) - \mu(t_0)\,x(t_0)
=\int_{t_0}^{t}\mu(s)\,q(s)\,ds.
$$

Como $\mu(t_0)=e^{\int_{t_0}^{t_0}p}=1$, obtenemos

$$
x(t)
=\mu(t)^{-1}\,x_0
+\mu(t)^{-1}\int_{0}^{t}\mu(s)\,q(s)\,ds.
$$

En nuestro caso $p(t)=2$, luego

$$
\int_{0}^{s}p(\tau)\,d\tau
=\int_{0}^{s}2\,d\tau
=2s,
$$

y por tanto

$$
\mu(s)=e^{2s},
\qquad
\mu(t)=e^{2t}.
$$

---

### Paso 3. Simplificar el integrando particular

Necesitamos  

$$
\mu(s)\,q(s)
=e^{2s}\bigl(140+20e^{-2s}\bigr)
=140\,e^{2s}+20.
$$

Aquí se ven dos aportes:

- **$140\,e^{2s}$**: responde al término constante $140$.  
- **$20$**: surge de $e^{2s}\cdot20e^{-2s}=20$.

---

### Paso 4. Integrar término a término

$$
\int_{0}^{t}\bigl[140\,e^{2s}+20\bigr]\,ds
=140\int_{0}^{t}e^{2s}ds + 20\int_{0}^{t}1\,ds.
$$

1. $\displaystyle\int_{0}^{t}e^{2s}ds
=\Bigl[\tfrac{1}{2}e^{2s}\Bigr]_{0}^{t}
=\tfrac12\bigl(e^{2t}-1\bigr).$
   Entonces  
   $$
   140\int_{0}^{t}e^{2s}ds
   =140\cdot\tfrac12\bigl(e^{2t}-1\bigr)
   =70\bigl(e^{2t}-1\bigr).
   $$

2. $\displaystyle\int_{0}^{t}1\,ds = t$, luego  
   $$
   20\int_{0}^{t}1\,ds =20t.
   $$

Sumando obtenemos

$$
\int_{0}^{t}\mu(s)q(s)\,ds
=70\bigl(e^{2t}-1\bigr)+20t.
$$

---

### Paso 5. Construir la solución general

Recordando

$$
x(t)
=e^{-\int_{0}^{t}p}\,x_0
+e^{-\int_{0}^{t}p}\int_{0}^{t}\mu(s)q(s)\,ds,
$$

y que $\int_{0}^{t}p=2t$, obtenemos

$$
x(t)
=e^{-2t}\cdot40
+e^{-2t}\bigl[70(e^{2t}-1)+20t\bigr].
$$

---

### Paso 6. Paso a paso de la simplificación algebraica

1. Expandimos el segundo término:

   $$
   e^{-2t}\bigl[70e^{2t}-70+20t\bigr]
   =70-70e^{-2t}+20t\,e^{-2t}.
   $$

2. Sumamos con $40e^{-2t}$:

   $$
   x(t)
   =40e^{-2t}+70-70e^{-2t}+20t\,e^{-2t}.
   $$

3. Agrupamos los exponentes:

   $$
   40e^{-2t}-70e^{-2t}=(40-70)e^{-2t}=-30e^{-2t}.
   $$

4. Resultado intermedio:

   $$
   x(t)=70-30e^{-2t}+20t\,e^{-2t}.
   $$

5. Factorizamos de modo que quede claro:

   $$
   x(t)=70+(20t-30)\,e^{-2t}.
   $$

---

### Paso 7. Interpretación cualitativa del resultado

- **Solución estacionaria**: $\displaystyle\lim_{t\to\infty}x(t)=70$. Esto coincide con resolver $-2x+140=0$.  
- **Transitorio**: $(20t-30)e^{-2t}$ decae a cero, pues $e^{-2t}$ domina para $t\to\infty$, aun cuando $20t$ crece linealmente.  
- En $t=0$, el transitorio vale $(0-30)\cdot1=-30$, compensando para que $x(0)=70-30=40$.

---

### Paso 8. Verificación de la condición inicial y de la ecuación

1. **Condición inicial**:

   $$
   x(0)=70+(20\cdot0-30)\,e^{0}=70-30=40.
   $$

2. **Comprobación de la EDO**:  
   - Derivada directa:

     $$
     x'(t)
     =\frac{d}{dt}\bigl[70+(20t-30)e^{-2t}\bigr]
     =20e^{-2t} + (20t-30)(-2e^{-2t})
     =(20-40t+60)e^{-2t}
     =(80-40t)\,e^{-2t}.
     $$

   - Sustituyendo en $-2x+140+20e^{-2t}$:

     $$
     -2\bigl[70+(20t-30)e^{-2t}\bigr]
       +140+20e^{-2t}
     =( -140 -40t\,e^{-2t}+60e^{-2t} )+140+20e^{-2t}
     =(80-40t)e^{-2t},
     $$

     que coincide con $x'(t)$.  

Por tanto, **la solución** del problema de valor inicial es

$$
\boxed{
x(t)=70 + (20t - 30)\,e^{-2t}
}
$$
